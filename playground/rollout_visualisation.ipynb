# Visualising rollouts

## Setup

```python
WANDB_PROJECT = "pvg-sandbox"
WANDB_RUN_NAME = "ppo_gi_next_norm_hist_no_round_1_1"
ITERATION = 9999
```

```python
from pvg.graph_isomorphism import GraphIsomorphismRolloutSamples
```

## Visualisation

```python
with GraphIsomorphismRolloutSamples(
    run_id=WANDB_RUN_NAME, iteration=ITERATION, wandb_project=WANDB_PROJECT
) as rollout_samples:
    rollout_samples.visualise()
    samples = rollout_samples._rollout_samples
```

```python
for i in range(10):
    print(samples[i]["y"], samples[i]["agents"]["decision"])
```

```python
for i in range(10):
    print(samples[i]["y"])
    print(samples[i]["agents"]["graph_level_repr"][0, 0])
# samples[0]["agents"]["graph_level_repr"][0, 1]
```
