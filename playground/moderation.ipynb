```python
from openai import OpenAI
import json
from tqdm import tqdm

client = OpenAI()
```

```python
file_path = '../data/code_validation/fine_tune_dataset(4).jsonl'

# Read the JSONL file
with open(file_path, 'r') as file:
    data = [json.loads(line) for line in file]
```

```python
flags = {}

for i in tqdm(range(len(data))):

  for m in data[i]['messages']:

    response = client.moderations.create(
      model="omni-moderation-latest",
      input=m['content'],
    )

    if response.results[0].flagged:

      flags[i] = {"role": m['role'],
                  "content": m['content'],
                  "response": response.model_dump_json()}

      print(f"Flagged: datum {i}")
      # categories = ", ".join([c for c in response['results']['categories'].keys() if response['results']['categories'][c]])
      # print(f"Flagged Categories: {categories}")
```

```python
# Save the flags dictionary as a JSONL file
output_file_path = '../data/code_validation/flags.jsonl'

with open(output_file_path, 'w') as outfile:
  for key, value in flags.items():
    json.dump( {"number": key,
                "role": flags[key]["role"],
                "content": flags[key]["content"],
                "response": json.loads(flags[key]["response"].model_dump_json())}
                    , outfile)
    outfile.write('\n')
```

```python
# Read the JSONL file
with open(output_file_path, 'r') as file:
    output_data = [json.loads(line) for line in file]

for d in output_data:
  print(d['number'])
  print(d["content"])
  for k in d["response"]["results"][0]["categories"].keys():
    if d["response"]["results"][0]["categories"][k]:
      print(k)
  print("\n\n\n\n")
```
